# Regression Analysis – Validation Notebook
This notebook demonstrates the `LinearRegressionStatisticalView` class and the `compare` function.

```python
import pandas as pd
import numpy as np
from linear_regression_view import LinearRegressionStatisticalView, compare
```

## 1. Load Dataset
We use the built-in `mtcars`-style dataset available from a public CSV (Auto MPG dataset).

```python
df = pd.read_csv(r"C:/Users/Silvana Marzouk/Downloads/mpg.csv")
df.dropna(subset=["mpg", "horsepower", "weight"])
df.head()
```

## 2. Choose Variables
- **Response (Y):** `mpg` (miles per gallon)  
- **Predictor 1 (X₁):** `horsepower`  
- **Predictor 2 (X₂):** `weight`

```python
X1 = df["horsepower"].values
X2 = df["weight"].values
Y  = df["mpg"].values

print(f"Number of observations: {len(Y)}")
```

## 3. Fit Model on Predictor 1 (horsepower)

```python
model = LinearRegressionStatisticalView(alpha=0.05)
model.fit(X1, Y)
```

## 4. Summary Output

```python
summary = model.summary()

print("=== Model Summary (horsepower ~ mpg) ===")
print(f"  F-statistic  : {summary['f_stat']:.4f}")
print(f"  F-critical   : {summary['f_critical']:.4f}")
print(f"  R²           : {summary['r_squareddef']:.4f}")
print(f"  R (Pearson)  : {summary['r']:.4f}")
print(f"  Degrees of freedom: {summary['degrees_of_freedom']}")
print(f"  Beta CIs     : {summary['beta_confidence_intervals']}")
```

## 5. Compare Two Predictors

```python
result = compare(df, x_col_1="horsepower", x_col_2="weight", y_col="mpg", alpha=0.05)

print("=== Comparison Result ===")
for k, v in result.items():
    print(f"  {k}: {v}")
```

## 6. Interpretation

The `compare` function fits two separate simple linear regression models predicting `mpg`:  
one using `horsepower` and one using `weight`.

The **F-statistic** measures how much of the variance in `mpg` is explained by each predictor relative to unexplained variance. A higher F-statistic means the predictor has stronger explanatory power.

Based on the output above, the `best_predictor` field identifies whichever variable yields the larger F-statistic — meaning it provides a significantly better linear fit for fuel efficiency (`mpg`) than the other variable.
